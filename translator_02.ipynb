```python
import pandas as pd
```

```python
from googletrans import Translator

translater = Translator()
```

```python
df = pd.read_excel("cleaned_reviews_data.xlsx")
```

```python
# getting 200 to 300 reviews from  the data set to avoid api blocking.

nw_df = df.loc[5701:6000,:]
```

```python
nw_df.shape
```

```python
# Creat a new column ("reviews") that contains translated Reviews of Portuguese

def translate_to_english(text):
  
  text2 = str(text)
  trans = translater.translate(text2, dest='en')
  return trans.text

nw_df["reviews_en"] =  pd.DataFrame(nw_df["reviews"].apply(lambda x: translate_to_english(x)))
```
